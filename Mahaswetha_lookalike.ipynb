```python
# Load the datasets
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')
```

```python
# Merge transactions with customers and products
transaction_data = transactions.merge(customers, on='CustomerID').merge(products, on='ProductID')
```

```python
# Create a customer profile with total spending and product categories
customer_profile = transaction_data.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'ProductID': lambda x: list(set(x)),  # Unique products purchased
    'Category': lambda x: list(set(x))     # Unique categories purchased
}).reset_index()
```
